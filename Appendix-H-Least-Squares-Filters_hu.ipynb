$$E = \sum_{j=0}^k |p(x_j) - y_j|^2$$

Feltételeztem, hogy a súlygyarapodásom állandó 1 font/nap, és így amikor megpróbáltam egy $n=1$ polinomot illeszteni, ami egy egyenes, az eredmény nagyon szorosan megfelelt a tényleges súlygyarapodásnak. De természetesen senki sem folyamatosan csak hízik vagy fogy. Ingadozunk. A 'polyfit()' használata hosszabb adatsorokhoz gyenge eredményeket adna. Ezzel szemben a g-h szűrő reagál az arány változásaira – a $h$ kifejezés szabályozza, hogy a szűrő milyen gyorsan reagál ezekre a változásokra. Ha hízunk, egy ideig stabilan tartjuk, majd lefogyunk, a szűrő automatikusan követi a változást.
