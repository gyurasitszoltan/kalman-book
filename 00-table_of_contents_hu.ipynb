[**Chapter 1: The g-h Filter**](./01-g-h-filter.ipynb)

Intuitív bevezetés a g-h szűrőhöz, más néven $\alpha$-$\beta$ szűrőhöz, amely a Kalman-szűrőt is magában foglaló szűrőcsalád. Miután megértette ezt a fejezetet, megérti a Kálmán-szűrő mögött meghúzódó fogalmakat. 


[**Chapter 2: The Discrete Bayes Filter**](./02-Discrete-Bayes.ipynb)

Bemutatja a diszkrét Bayes szűrőt. Ebből könnyen emészthető formában megtanulhatja a valószínűségi (bayesi) érvelést, amely a Kalman-szűrőt támasztja alá.

[**Chapter 3: Probabilities, Gaussians, and Bayes' Theorem**](./03-Gaussians.ipynb)

Bevezeti a Gauss-féle hiedelmek bayesi értelemben vett ábrázolását. A Gauss-féle algoritmusok lehetővé teszik számunkra, hogy a diszkrét Bayes-szűrőben használt algoritmusokat folytonos tartományokban működjenek.


[**Chapter 4: One Dimensional Kalman Filters**](./04-One-Dimensional-Kalman-Filters.ipynb)

Valósít egy Kalman-szűrőt a diszkrét Bayes-szűrő módosításával a Gauss-féle használathoz. Ez egy teljes értékű Kalman szűrő, bár csak 1D problémák esetén hasznos. 


[**Chapter 5: Multivariate Gaussians**](./05-Multivariate-Gaussians.ipynb)

A Gauss-féle több dimenzióra kiterjeszti, és bemutatja, hogy a „háromszögelés” és a rejtett változók hogyan javíthatják jelentősen a becsléseket.

[**Chapter 6: Multivariate Kalman Filter**](./06-Multivariate-Kalman-Filters.ipynb)

Az egyváltozós fejezetben kifejlesztett Kálmán-szűrőt kiterjesztjük a lineáris problémák teljes, általánosított szűrőjére. Ennek elolvasása után meg fogja érteni, hogyan működik a Kalman-szűrő, és hogyan kell megtervezni és megvalósítani egyet az Ön által választott (lineáris) problémára.

[**Chapter 7: Kalman Filter Math**](./07-Kalman-Filter-Math.ipynb)

Körülbelül olyan messzire jutottunk, amennyire csak lehet, anélkül, hogy erős matematikai alapot alkottunk volna. Ez a fejezet nem kötelező, különösen az első alkalommal, de ha robusztus, numerikusan stabil szűrőket szeretne írni, vagy olvasni kívánja a szakirodalmat, akkor ismernie kell a fejezet anyagát. Néhány szakaszra szükség lesz a nemlineáris szűrésről szóló későbbi fejezetek megértéséhez. 


[**Chapter 8: Designing Kalman Filters**](./08-Designing-Kalman-Filters.ipynb)

Az 5. és 6. fejezetben található anyagokra építve végigvezeti Önt számos Kálmán-szűrő kialakításán. Csak több különböző példa láttán lehet igazán megérteni az egész elméletet. A példákat úgy választottuk ki, hogy valósághűek legyenek, nem pedig „játék” problémákat, hogy elinduljon a saját szűrőinek megvalósítása felé. Megvitatja, de nem oldja meg az olyan kérdéseket, mint a numerikus stabilitás.


[**Chapter 9: Nonlinear Filtering**](./09-Nonlinear-Filtering.ipynb)

A Kálmán szűrők, ahogyan a leírásban szerepel, csak lineáris problémák esetén működnek. A világ azonban nemlineáris. Itt bemutatom a nemlineáris rendszerek által a szűrővel kapcsolatos problémákat, és röviden tárgyalom a különféle algoritmusokat, amelyeket a következő fejezetekben fogunk megtanulni.


[**Chapter 10: Unscented Kalman Filters**](./10-Unscented-Kalman-Filter.ipynb)

Az illatmentes Kalman-szűrők (UKF) a Kalman-szűrőelmélet legújabb fejlesztései. Lehetővé teszik a nemlineáris problémák szűrését anélkül, hogy olyan zárt formájú megoldásra lenne szükség, mint az Extended Kalman szűrőnél.

Ezt a témát általában vagy nem említik, vagy elfedik a meglévő szövegek, és a kiterjesztett Kálmán-szűrők kapják a vita nagy részét. Azért tettem az első helyre, mert az UKF-et sokkal egyszerűbb megérteni, megvalósítani, és a szűrési teljesítmény általában olyan jó vagy jobb, mint az Extended Kalman szűrőé. Mindig igyekszem először az UKF-et megvalósítani a való világ problémáira, és neked is meg kell tenned.


[**Chapter 11: Extended Kalman Filters**](./11-Extended-Kalman-Filters.ipynb)

A kiterjesztett Kalman-szűrők (EKF) a legelterjedtebb megközelítés a nemlineáris problémák linearizálására. A valós Kálmán-szűrők többsége EKF, ezért meg kell értenie ezt az anyagot, hogy megértse a meglévő kódot, papírokat, beszélgetéseket stb. 


[**Chapter 12: Particle Filters**](./12-Particle-Filters.ipynb)

A részecskeszűrők Monte Carlo technikákat használnak az adatok szűrésére. Könnyen kezelik az erősen nemlineáris és nem Gauss-rendszereket, valamint a multimodális eloszlásokat (több objektum egyidejű nyomon követése), magas számítási követelmények árán.


[**Chapter 13: Smoothing**](./13-Smoothing.ipynb)

A Kalman szűrők rekurzívak, így nagyon alkalmasak valós idejű szűrésre. Az adatok utófeldolgozásánál azonban rendkívül jól működnek. Hiszen a Kálmán-szűrők prediktor-javítók, és könnyebb megjósolni a múltat, mint a jövőt! Megbeszélünk néhány közös megközelítést.


[**Chapter 14: Adaptive Filtering**](./14-Adaptive-Filtering.ipynb)
    
A Kalman-szűrők egyetlen folyamatmodellt feltételeznek, de a manőverezési célokat általában több különböző folyamatmodellel kell leírni. Az adaptív szűrés számos technikát használ annak érdekében, hogy a Kalman-szűrő alkalmazkodjon a cél változó viselkedéséhez.
