## SymPy

A SymPy egy Python-csomag szimbolikus matematika végrehajtására. Képességeinek teljes skálája túlmutat ezen a könyvön, de képes algebrát végezni, egyenleteket integrálni és differenciálni, megoldásokat találni differenciálegyenletekre és még sok mást. Használjuk például a mátrixok Jacobi-számítására és a várható értékű integrálszámításokra.

Először is egy egyszerű példa. Importáljuk a SymPy-t, inicializáljuk a szép nyomtatási funkcióját (amely egyenleteket nyomtat a LaTeX használatával). Ezután deklarálunk egy szimbólumot a SymPy számára.

```python
import sympy
sympy.init_printing(use_latex='mathjax')

phi, x = sympy.symbols('\phi, x')
phi
```

Figyeld meg, hogyan nyomtatja ki a „phi” szimbólumot a LaTeX használatával. Most számoljunk egy kicsit. Mi a $\sqrt{\phi}$ deriváltja?

```python
sympy.diff('sqrt(phi)')
```

Tényezős egyenleteket tudunk adni

```python
sympy.factor(phi**3 -phi**2 + phi - 1)
```

és bővíthetjük őket.

```python
((phi+1)*(phi-4)).expand()
```

Kiértékelhet egy egyenletet a változói adott értékeire:

```python
w =x**2 -3*x +4
print(w.subs(x, 4))
print(w.subs(x, 12))
```

Használhat karakterláncokat olyan egyenletekhez is, amelyek olyan szimbólumokat használnak, amelyeket nem definiált:

```python
x = sympy.expand('(t+1)*2')
x
```

Most a SymPy segítségével számítsuk ki egy mátrix Jacobi-át. Adott a funkció

$$h=\sqrt{(x^2 + z^2)}$$

keresse meg a Jacobi-t x, y és z vonatkozásában.

```python
x, y, z = sympy.symbols('x y z')

H = sympy.Matrix([sympy.sqrt(x**2 + z**2)])

state = sympy.Matrix([x, y, z])
H.jacobian(state)
```

Most számítsuk ki a diszkrét folyamatzajmátrixot $\mathbf Q$ a folytonos folyamatzajmátrix alapján 
$$\mathbf Q = \Phi_s \begin{bmatrix}0&0&0\\0&0&0\\0&0&1\end{bmatrix}$$

Az integrál az 

$$\mathbf Q = \int_0^{\Delta t} \mathbf F(t)\mathbf Q\mathbf F^T(t)\, dt$$

hol 
$$\mathbf F(\Delta t) = \begin{bmatrix}1 & \Delta t & {\Delta t}^2/2 \\ 0 & 1 & \Delta t\\ 0& 0& 1\end{bmatrix}$$

```python
dt = sympy.symbols('\Delta{t}')
F_k = sympy.Matrix([[1, dt, dt**2/2],
                    [0,  1,      dt],
                    [0,  0,       1]])
Q = sympy.Matrix([[0,0,0],
                  [0,0,0],
                  [0,0,1]])

sympy.integrate(F_k*Q*F_k.T,(dt, 0, dt))
```
